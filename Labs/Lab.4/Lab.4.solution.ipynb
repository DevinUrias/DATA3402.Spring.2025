1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
import math
class Counter:
    def __init__(self, maximum):
        self.maximum=maximum
        self.count=0
        
    def increment(self):
        if self.count < self.maximum:
            self.count += 1
        else:
            print("Error: Counter has reached its maximum amount.")
        
    def reset(self):
        print("Counter has been reset.")
        self.count = 0
    
    def display_count(self):
        return self.count
            
```

```python

counter = Counter(5)

for _ in range(5):
    counter.increment()
    print("Counter Value:", counter.display_count())

counter.increment()
counter.count=3
counter.increment()
print("Counter Value:", counter.display_count())
counter.reset()
print("Counter Value:", counter.display_count())

```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self):
        self.shapes = dict()
        self.shape_names = list()
    
    def add_shape(self, shape, name=None):
        if not name:
            name = self.assign_name()
        shape.name = name
        self.shapes[name] = shape
        self.shape_names.append(name)
    
    def remove_shape(self, name):
        if name in self.shapes:
            del self.shapes[name]
            self.shape_names.remove(name)
    
    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)
        
    def paint(self,canvas):
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(canvas)
            
    def assign_name(self):
        name_base = "shape"
        i = 0
        while f"{name_base}_{i}" in self.shapes:
            i += 1
        return f"{name_base}_{i}"
```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
